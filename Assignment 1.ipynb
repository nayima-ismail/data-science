**Que.7 What is the output of the following code?**   

**Ans** This is a nested loop working with 2 variable i an j.  

range() is used to generate a sequence of numbers.  
It takes only whole number input. does not work on floats. 
range(start, stop, step), start cannot be less stop unless the step is a negative number. Then range is generated in descending order. 

|i |j | print(i)|
|--|--|---|
|1 |2 | prints nothing as a range(2, 1) start< stop, nested loop not executed|
|2 |2 | default step size is 1. range(2,2) has start=stop, nested loop not executed|
|3 |2 | loop executes in range(2,3): loops executes once, i%j -> 2%2!= 1 hence does not print anything|
|4 |2, 3 | loop executes twice, prints 3 for 4%3 == 1 |
|15|2 - 14 | range(1,16) starts at 1 and stops at 15 |

```python
for i in range(1, 16): 
    for j in range(2, i): 
        if i%j == 1: 
            print(i) 
            break
```

**Que. 10 Generate two 4x4 2D numpy array using arrange(), and then perform some basic operation and evaluate the results.**  
  
    a. Transpose the array  
    b. Multiply the arrays  
    c. Stack the arrays Vertically and Horizontally  
    d. Convert the 2D array into 1D array   
    e. Reverse the column of the array   
    f. Swap two rows of an array   
    g. Swap two columns of an array   
    h. Get the common elements between two arrays   
    i. Get the position of if two array matches somewhere.   
    j. Remove the elements of an array if it matches to another array.  

**Ans:**  arange(start, stop, step, dtype, like) returns a ndarray type
 default value of start 0, optional
 stop is mandatory
 step size is 1 by default
 dtype data type of items in the list
 like 

```python
import numpy as np
rangeA = np.arange(16)
rangeB = np.arange(-5, 11)
a = rangeA.reshape(4,4)
b = rangeB.reshape(4,4)
print("Original 4X4 2D arrays \n",a,"\n", b)

# a. Transpose the array  
print("Transpose the array  :\n", np.transpose(a), "\n", np.transpose(b),"\n")
#b. Multiply the arrays  array  
print("Multiplication of arrays  :\n", np.matmul(a,b))
#c. Stack the arrays Vertically and Horizontally  
print("Stack the arrays Horizontally :\n", np.hstack((rangeB, rangeA)), "\n")
print("Stack the arrays Vertically :\n", np.stack((rangeA,rangeB), axis=0),"\n")
#d. Convert the 2D array into 1D array   
print("Convert the 2D array into 1D array  :\n", np.ravel(a), "\n", np.ravel(b),"\n")
#e. Reverse the column of the array   
print("Reverse the column of the array  :\n", np.fliplr(a), "\n", np.fliplr(b))

#g. Swap two columns of an array   
a[:, [2, 0]] = a[:, [0, 2]]
print("Swap two columns of an array  :\n", a)
#f. Swap two rows of an array  
a[[0, 2]] = a[[2, 0]]
print("Swap two rows of an array  :\n", a)
intersection,a_ind, b_ind = np.intersect1d(a,b, return_indices=True)
#h. Get the common elements between two arrays   
print("Get the common elements between two arrays   :\n", intersection)
#i. Get the position of if two array matches somewhere.   
print("Get the position of if two array matches somewhere.  :\n", a_ind,"\n", b_ind)
#j. Remove the elements of an array if it matches to another array
print("Remove the elements of an array if it matches to another array.  :\n", np.delete(a, intersection))
```

**Que. 11 Take an image of your choice. Performs basic operations using SciPy library:**

    g. Apply morphological dilation and erosion with the variation of size: tuple of ints on the image.  

```python
from scipy import misc, ndimage
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# setting values to rows and column variables
rows = 4
columns = 3

# a
bloom = mpimg.imread('C:\\Users\\nismai01\\Downloads\\bloom.jpg') 

#b
plt.imshow(bloom)
plt.title("Bee")
plt.show()

#c
rotated_image = ndimage.rotate(bloom,30)
plt.title("Rotated Bee")
plt.imshow(rotated_image)
plt.show()


```

```python
#d Flipped Image
import numpy as np
flipped_image = np.flipud(bloom)
plt.title("Flipped Image")
plt.imshow(flipped_image)
plt.show()
```

```python
#e Sobel Filter
import scipy.ndimage.filters as filters

sobel_image = filters.sobel(bloom)
plt.imshow(sobel_image)
plt.show()
```

```python
#f
gauss_image_1 = filters.gaussian_filter(bloom, sigma=1)
gauss_image_2 = filters.gaussian_filter(bloom, sigma=2)
gauss_image_3 = filters.gaussian_filter(bloom, sigma=3)
gauss_image_4 = filters.gaussian_filter(bloom, sigma=4)
gauss_image_5 = filters.gaussian_filter(bloom, sigma=5)

plt.title("Sigma 1")
plt.imshow(gauss_image_1)
plt.show()
plt.title("Sigma 2")
plt.imshow(gauss_image_2)
plt.show()
plt.title("Sigma 3")
plt.imshow(gauss_image_3)
plt.show()
plt.title("Sigma 4")
plt.imshow(gauss_image_4)
plt.show()
plt.title("Sigma 5")
plt.imshow(gauss_image_5)
plt.show()
```

```python
#g
import scipy.ndimage.morphology as morph

# had to convert to grayscale or else throws error tuple out of index
from skimage.color import rgb2gray
grey_image = rgb2gray(bloom) 

grey_dilation = morph.grey_dilation(grey_image, size=(3,3), structure=np.ones((3, 3)))
plt.title("Grey Dilation")
plt.imshow(grey_dilation)
plt.show()

grey_erosion= morph.grey_erosion(grey_image, size=(100,50), origin=1)
plt.title("Grey Erosion")
plt.imshow(grey_erosion)
plt.show()
```
