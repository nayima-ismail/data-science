```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import and plot the dataset
from sklearn.datasets import make_moons

Xmoon, Ymoon = make_moons(300, noise=.08, random_state=0)
plt.scatter(Xmoon[:, 0], Xmoon[:, 1]);
```

```python
Ymoon
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=0)
```

```python
gmm.fit(Xmoon)
labels = gmm.predict(Xmoon)
plt.scatter(Xmoon[:, 0], Xmoon[:, 1], c=labels, s=40, cmap='viridis')
```

```python
newSample = gmm.sample(100)
newSample
```

```python
Xnew=newSample[0]
plt.scatter(Xnew[:, 0], Xnew[:, 1]);
```

```python
newlabels = gmm.predict(Xnew)
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=newlabels, s=40, cmap='viridis')
# plt.scatter(Xmoon[:, 0], Xmoon[:, 1], c=labels, s=40, cmap='cool')
```

```python
###Real Example:we will generate new images from the standard LSW corpus that we have used before.
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

```python
def plot_faces(faces):
    fig, ax = plt.subplots(3, 5)
    for i, axi in enumerate(ax.flat):
        axi.imshow(faces.images[i], cmap='bone')
        axi.set(xticks=[], yticks=[],
                xlabel=faces.target_names[faces.target[i]])
```

```python
faces
```

```python
plot_faces(faces)
```

```python
face_gmm = GaussianMixture(n_components=7, covariance_type='full', random_state=0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.99)
data = pca.fit_transform(faces.data)
data.shape
```

```python
#CLuster the data:
face_gmm.fit(data)
print(face_gmm.converged_)
```

```python
#Generating 100 new points:
new_faces = face_gmm.sample(100)
new_faces[0].shape
```

```python
#Applying inverse-PCA and plotting the data:
face = pca.inverse_transform(new_faces[0])
face
```
